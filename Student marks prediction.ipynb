```python
# Step 1: Import the necessary libraries
# pandas: for data manipulation
# numpy: for numerical operations
# matplotlib.pyplot: for creating visualizations
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 2: Create the dataset
# We define a dictionary with study hours and corresponding marks,
# then convert it into a pandas DataFrame for easier handling.
data = {
    "Hours" : [2,3,4,5,6,7,8,9,10],
    "Marks" : [35,45,55,65,75,82,88,92,98]
}
df = pd.DataFrame(data)
print(df)
```

```python
# Step 3: Visualize the data
# We separate the columns into x (input) and y (output) and create a plot
# to see the relationship between study hours and marks.
x = df['Hours']
y = df['Marks']
plt.xlabel("Hours")
plt.ylabel("Marks")
plt.plot(x, y) # Draws the line
plt.scatter(x, y) # Draws the individual data points
plt.show()
```

```python
# Step 4: Split the data into Training and Testing sets
# train_test_split helps us set aside 20% of the data to test the model later,
# while using the other 80% to train it.
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print(x_train)
print("Break")
print(x_test)
```

```python
print(y_train)
print("Break")
print(x_test)
```

```python
# Step 5: Initialize and Train the Linear Regression Model
# We create the model and fit it to our training data.
# Note: We use .reshape(-1, 1) because the model expects a 2D array for features (x).
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train.values.reshape(-1, 1), y_train)
```

```python
# Step 6: View the Model Parameters
# W (coef_) is the slope: how much Marks increase per 1 hour of study.
# B (intercept_) is the starting point: predicted Marks if Hours were zero.
print("W (Slope):", model.coef_)
print("B (Intercept):", model.intercept_)
```

```python
# Step 7: Make a Prediction
# We use the trained model to predict the Marks for a student who studies 11 hours.
prediction = model.predict([[11]])
print(f"Predicted Marks for 11 hours: {prediction[0]}")
```

```python
# Step 8: Evaluate the Model

# We assess the model's performance using R2 Score (how well predictions match actuals)
# and Mean Squared Error (average squared difference between predicted and actual values).
# A higher R2 and lower MSE indicate a better model fit.


from sklearn.metrics import r2_score, mean_squared_error

y_pred = model.predict(x_test.values.reshape(-1, 1))

print("R2 Score:", r2_score(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
```
