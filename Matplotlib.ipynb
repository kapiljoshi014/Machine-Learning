```python
#Use to import libary
import matplotlib.pyplot as plt
```

```python
#Basic plot
x=[1,2,3,4,5]
y=[10,20,30,40,50]
plt.xlabel("X axis")    #Use to give name on horizontal side(--->)
plt.ylabel("Y axis")    #Use to give name on verical side(|>)
plt.title("Basic plot") #use to give name
plt.plot(x,y)           #Draws it in memory,,but in colab, it will automatically show without using show().
plt.show()              #Displays it on the screen.
```

```python
#Scatter plot  is used to find relationships, patterns, or trends between two numerical variables.like,house size vs price.
plt.scatter(x,y)
plt.show()
```

```python
#Histogram used to show the distribution of a single numerical variable
ages=[10,20,30,40,50]
plt.hist(ages)
plt.show()
```

```python
#Bar Chart used to compare the values of different categories or groups
movies=["Sci-fi","thriller","horror","comedy"]
votes=[12,33,47,29]
plt.bar(movies,votes)
plt.show()
```
